```python
import os
import pandas as pd

# Set your spectrogram folder
spectrogram_dir = "/content/drive/MyDrive/audio_dataset/spectrograms_final1"

# Emotion code to label mapping
emotion_map = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

data = []

for fname in os.listdir(spectrogram_dir):
    if not fname.endswith(".png"):
        continue

    try:
        parts = fname.split("_")  # Example: Actor_01_03-01-01-01-01-01-01_speech.png
        actor_id = parts[1]
        emotion_code = parts[2].split("-")[2]
        emotion = emotion_map.get(emotion_code, "unknown")
        gender = "male" if int(actor_id) % 2 != 0 else "female"

        data.append({
            "filepath": os.path.join(spectrogram_dir, fname),
            "emotion_code": emotion_code,
            "emotion": emotion,
            "actor": actor_id,
            "gender": gender
        })

    except Exception as e:
        print(f"Error processing file {fname}: {e}")

# Create DataFrame and save
df = pd.DataFrame(data)
df.to_csv("/content/spectrogram_labels.csv", index=False)
print("✅ Saved spectrogram-label CSV:", df.shape)
```
